```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from matplotlib import pyplot as plt
import seaborn as sns
data = pd.read_csv("heart_v2.csv")
data.head()
```

```python
data.info()
# we have 5 columns all int type (numerical)
```

```python
data.describe()
#we can see there is no null value in any of columns
```

```python
# see how age is distributed
plt.hist(data["age"])
plt.show()
# most people here belongs to age 55-65
```

```python
# see how sex is distributed
plt.hist(data["sex"])
plt.show()
# 1:2 ratio on male:female
```

```python
# see how BP is distributed
plt.hist(data["BP"])
plt.show()
# most around 110-140
```

```python
# see how cholestrol is distributed
plt.hist(data["cholestrol"])
plt.show()
# most around 250
```

```python
# check how our target variable has it's values
data["heart disease"].value_counts()
# 150 yes, 120: no
```

```python
# make seperate x and y
x = data.drop("heart disease",axis=1)
y = data["heart disease"].copy()
x.head()
```

```python
y.head()
```

# Train test split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x , y, train_size=.7,random_state=40)
x_train.shape,y_train.shape
```

# Build the decision tree


```python
# now try with default parameters except depth of tree
dt = DecisionTreeClassifier(max_depth=3)
dt.fit(x_train,y_train)
```

```python
# visulize the decision tree
from sklearn.tree import plot_tree
plt.figure(figsize=(60,30))
plot_tree(dt, feature_names = x.columns, class_names = ["No Disease", "Disease"], filled = True)
plt.show()
```

# Evaluate model

```python
# make prediction over both train and test first.
y_train_pred = dt.predict(x_train)
y_test_pred = dt.predict(x_test)

from sklearn.metrics import confusion_matrix, accuracy_score
print(accuracy_score(y_train,y_train_pred))
print(confusion_matrix(y_train,y_train_pred))
# getting 73% accuracy on train set

print(accuracy_score(y_test,y_test_pred))
print(confusion_matrix(y_test,y_test_pred))
# getting 69% accuracy on test set, which got down to train set
```

```python
# now try with default parameters
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
# visulize the decision tree
from sklearn.tree import plot_tree
plt.figure(figsize=(60,30))
plot_tree(dt, feature_names = x.columns, class_names = ["No Disease", "Disease"], filled = True)
plt.show()
```

```python
# make prediction over both train and test first.
y_train_pred = dt.predict(x_train)
y_test_pred = dt.predict(x_test)

from sklearn.metrics import confusion_matrix, accuracy_score
print(accuracy_score(y_train,y_train_pred))
print(confusion_matrix(y_train,y_train_pred))
# getting 100% accuracy on train set

print(accuracy_score(y_test,y_test_pred))
print(confusion_matrix(y_test,y_test_pred))
# getting 62% accuracy on test set, which got down to train set
```

# hyper parameter tuning using Grid search CV

```python
dt = DecisionTreeClassifier(random_state=42)
from sklearn.model_selection import GridSearchCV
params = {
    'max_depth': [2,3,5,10,20],
    'min_samples_leaf' : [5,10,20,50,100],
    'criterion' : ['gini', 'entropy']
}
```

```python
grid_search = GridSearchCV(estimator=dt,
            param_grid=params,
            cv=4,
            n_jobs=-1,verbose=1,
            scoring="accuracy")
```

```python
%%time
grid_search.fit(x_train,y_train)
```

```python
# take in DF and show
df_search = pd.DataFrame(grid_search.cv_results_)
df_search.head()
```

```python
print(len(df_search))
df_search.sort_values(by = "rank_test_score").head()
# rank_test_score =1 we get .68 == 68% as mean score on params entropy, depth = 10 and leaf = 5
```

```python
# best params
grid_search.best_params_
```

```python
# best score (mean)
grid_search.best_score_
```

```python
dt_best = grid_search.best_estimator_
```

```python
# visulize the decision tree on these best score
from sklearn.tree import plot_tree
plt.figure(figsize=(60,30))
plot_tree(dt_best, feature_names = x.columns, class_names = ["No Disease", "Disease"], filled = True)
plt.show()
```

```python
# make prediction over both train and test first.
y_train_pred = dt_best.predict(x_train)
y_test_pred = dt_best.predict(x_test)

from sklearn.metrics import confusion_matrix, accuracy_score
print(accuracy_score(y_train,y_train_pred))
print(confusion_matrix(y_train,y_train_pred))
# getting 100% accuracy on train set

print(accuracy_score(y_test,y_test_pred))
print(confusion_matrix(y_test,y_test_pred))
# getting 62% accuracy on test set, which got down to train set
```
