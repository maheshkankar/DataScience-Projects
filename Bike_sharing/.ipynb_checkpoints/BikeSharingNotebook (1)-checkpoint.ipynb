```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("day.csv")
data.head()
```

```python
data.info()
# dteday is only object rest numerical variables
```

```python
data.describe()
```

```python
data.shape
# there are 730 rows and 16 columns in dataset
```

```python
data.columns
# these are folling variables
```

```python
# check for null or missing values
data.isnull().sum()
```

```python
# check which columns can be removed
data[["dteday","mnth","yr","casual" ,"registered"]].head()
# dtedate gives info which we can have from month and yr so can remove it.
# also casual and registered gives info of catagories which is not aligned to our objecive so will remove it.
```

```python
data["instant"].value_counts() # it's only inde value so can remove it
```

```python
data.drop(["dteday","instant","casual" ,"registered"],inplace = True, axis =1)
data.head()
```

```python
# take old backup 
data_old = data
```

```python
# as we can see we have 'mnth', 'weekday', 'season' & 'weathersit' variables which are categriocal so need DUMMY for them.

# But these are int or flot let's convert them into 'category' data types first.
data['season']=data['season'].astype('category')
data['weathersit']=data['weathersit'].astype('category')
data['mnth']=data['mnth'].astype('category')
data['weekday']=data['weekday'].astype('category')
data.info()
```

```python
# create dummy
data = pd.get_dummies(data, drop_first=True)
data.info()
```

```python
data.shape
# now we have total 30 columns
```

```python
# perform some EDA
# for numeric variables 
data_num=data[['temp', 'atemp', 'hum', 'windspeed','cnt']]

sns.pairplot(data, diag_kind='kde')
plt.show()
```

```python
#to see how each of the predictor variable related against the target variable.

plt.figure(figsize=(25, 10))
plt.subplot(2,3,4)
sns.boxplot(x = 'holiday', y = 'cnt', data = data_old)
plt.subplot(2,3,5)
sns.boxplot(x = 'weekday', y = 'cnt', data = data_old)
plt.subplot(2,3,6)
sns.boxplot(x = 'workingday', y = 'cnt', data = data_old)
plt.subplot(2,3,1)
sns.boxplot(x = 'season', y = 'cnt', data = data_old)
plt.subplot(2,3,2)
sns.boxplot(x = 'mnth', y = 'cnt', data = data_old)
plt.subplot(2,3,3)
sns.boxplot(x = 'weathersit', y = 'cnt', data = data_old)
plt.show()
#When not holyday sharing happened most
# Most bike sharing in month 5,6,7,8,9
# bike sharing more on workingday.
```

```python
# corelation 
plt.figure(figsize = (25,20))
sns.heatmap(data_old.corr(), annot = True, cmap="RdBu")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Start with modeling 
# split data into train and test

np.random.seed(0)
data_train, data_test = train_test_split(data, train_size = 0.70, test_size = 0.30, random_state = 333)
print(data_train.shape,data_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# apply scaling to numeric variables
num_vars = ['temp', 'atemp', 'hum', 'windspeed','cnt']

data_train[num_vars] = scaler.fit_transform(data_train[num_vars])
# Checking values after scaling
data_train.head()
```

```python
# split target and dependent
y_train = data_train.pop('cnt')
X_train = data_train
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python

# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = data.columns
vif['VIF'] = [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
