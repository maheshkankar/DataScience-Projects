```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import sklearn
import sklearn.preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
from sklearn import svm
```

```python
data = pd.read_csv("diabetes_data.csv")
data.head()
# data dictonary Reference
# https://www.kaggle.com/datasets/alexteboul/diabetes-health-indicators-dataset?select=diabetes_binary_health_indicators_BRFSS2015.csv
```

```python
data.info()
# no any column having null values
# all collumns are of numirical type
```

```python
data.describe()
```

```python
# let's have some feature engineering
# HighBP is binary
# 0 = no high BP 1 = high BP
plt.hist(data["HighBP"])
plt.show()
# more patients in dataset having no BP
```

```python
# BMI having range of valuees
plt.hist(data["BMI"])
plt.show()
# more patients in dataset having BMI range 25-30
```

```python
# Age 
plt.hist(data["Age"])
plt.show()
# age is encoded 
#13-level age category (_AGEG5YR see codebook) 1 = 18-24 9 = 60-64 13 = 80 or older
```

```python
# Income of patients is been categorized.
plt.hist(data["Income"])
plt.show()
# Income scale (INCOME2 see codebook) scale 1-8 1 = less than $10,000 5 = less than $35,000 8 = $75,000 or more
```

```python
# Splitting the dataset into training and testing sets.
y = data.pop("Diabetes_binary")
x = data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0, test_size = 0.2)
x_train.head()
```

```python
y_train.head()
```

```python
# Creating the SVM model.
clf = svm.SVC(kernel='rbf')
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```
